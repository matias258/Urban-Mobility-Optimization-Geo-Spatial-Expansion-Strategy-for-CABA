# Urban Mobility Optimization: Strategic Hub Localization
### Case Study: Fleet Positioning in Buenos Aires (CABA)

1. **Objective:** Identify strategic points for preventive driver positioning in Buenos Aires.
2. **Methodology:** Use the official taxi stops dataset (as a proxy for historical demand) and apply a **K-Means Clustering** algorithm to solve a **Location-Allocation** problem (Operations Research).

```python
import pandas as pd
import re
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import folium
import networkx as nx
from branca.element import Template, Element
import warnings
import networkx as nx
from scipy.spatial.distance import cdist
warnings.filterwarnings('ignore')
from sklearn.metrics.pairwise import haversine_distances
```

## 1. Data Preparation (ETL)
We extract geographical coordinates from the `geometry` column to enable vector operations with the city's data points.

```python
# Load the Buenos Aires taxi stops dataset
df = pd.read_csv('datasets/paradas_taxis.csv')

def extract_coords(geometry):
    # Original format is POINT (longitude latitude)
    coords = re.findall(r"[-+]?\d*\.\d+|\d+", geometry)
    return float(coords[1]), float(coords[0])

# Generate clean Latitude and Longitude columns
df[['lat', 'lon']] = df['geometry'].apply(lambda x: pd.Series(extract_coords(x)))

print(f"Total analyzed points: {len(df)}")
df[['barrio', 'lat', 'lon']].head()
```

## 2. Operations Research Modeling
We use **K-Means** to identify 10 "gravity centers" of existing supply.
This analysis minimizes the total distance between any taxi stop (demand) and the nearest Expansion Hub (supply), optimizing the projected **ETA (Estimated Time of Arrival)**.

```python
# Define 10 strategic clusters
K_HUBS = 10
kmeans = KMeans(n_clusters=K_HUBS, random_state=42, n_init=10)

# Assign each stop to a cluster and calculate centroids (Hubs)
df['cluster'] = kmeans.fit_predict(df[['lat', 'lon']])
hubs = kmeans.cluster_centers_

print("Strategic Hub coordinates successfully generated.")
```

## 4. Network Topology & Connectivity Analysis
Beyond spatial clustering, we model the expansion hubs as a **Graph**. This allows us to analyze the "inter-connectivity" of the city. 

- **Nodes:** Expansion Hubs (Strategic Centers).
- **Edges:** Feasible transition paths for drivers between hubs (set at a ~3km threshold).
- **Goal:** Identify the "backbone" of the mobility network and ensure fleet rebalancing is operationally viable.

```python
# Initialize a Graph to represent the 10 Strategic Hubs
G = nx.Graph()

# Add hubs as nodes with their geographical positions
for i, hub in enumerate(hubs):
    G.add_node(i, pos=(hub[1], hub[0])) # (longitude, latitude)

# Calculate Euclidean distance matrix between all hubs
dist_matrix = cdist(hubs, hubs, metric='euclidean')

# Establish connections (edges) based on a proximity threshold
# This simulates the ability of a driver to "jump" or rebalance between nearby hubs
THRESHOLD = 0.05 # Approximate degree threshold (~5km)

for i in range(len(hubs)):
    for j in range(i + 1, len(hubs)):
        if dist_matrix[i, j] < THRESHOLD:
            G.add_edge(i, j, weight=dist_matrix[i, j])
```

## Betweenness Centrality (Intermediación)
Este es el algoritmo que más valor aporta a un análisis de expansión logística. La Centralidad de Intermediación mide cuántas veces un nodo (Hub) actúa como puente para el camino más corto entre otros nodos.

## 5. Advanced Network Topology & Connectivity Analysis

While the previous clustering analysis (K-Means) identified the optimal centers of demand, real-world logistics require understanding how these centers interact. In this section, we transition from **spatial points** to a **weighted network system**.

### Operational Objectives:
- **Haversine Weighting ($W$):** Unlike simple Euclidean distance, we implement the **Haversine formula** to calculate real-world distances in kilometers, accounting for the Earth's curvature.
- **Betweenness Centrality:** We apply this Graph Theory algorithm to identify "bridge" hubs. A hub with high centrality is a critical articulation point where driver flow is most concentrated.
- **Supply Chain Resilience:** By visualizing the "backbone" of the network, we can identify which hubs are vital for fleet rebalancing and which act as **Satellite Outposts** (Isolated Nodes).

### Key Technical Findings:
1. **The Infrastructure Backbone:** The gray corridors represent the most efficient paths for a driver to move between strategic demand zones within a 5km threshold.
2. **Node Hierarchy:** Red nodes are sized proportionally to their **Centrality Score**. Larger nodes indicate high-traffic "pivots" (like Retiro or Constitución) that are essential for maintaining city-wide service levels.
3. **Strategic Isolation:** Any node without edges represents a self-contained market. These areas require a dedicated fleet strategy as they are geographically decoupled from the main rebalancing network.

```python
# 1. Calculate Real Distances (Haversine) and Centrality
# We convert to radians for the formula
hubs_in_radians = np.radians(hubs)
dist_matrix_km = haversine_distances(hubs_in_radians) * 6371

# Rebuild the weighted graph
G_weighted = nx.Graph()
for i, hub in enumerate(hubs):
    G_weighted.add_node(i, pos=(hub[0], hub[1]))

THRESHOLD_KM = 5.0 
for i in range(len(hubs)):
    for j in range(i + 1, len(hubs)):
        distance = dist_matrix_km[i, j]
        if distance < THRESHOLD_KM:
            G_weighted.add_edge(i, j, weight=distance)

# Calculate Importance (Betweenness Centrality)
centrality = nx.betweenness_centrality(G_weighted, weight='weight')

# 2. Initialize Map
mapa_caba = folium.Map(location=[-34.6037, -58.3816], zoom_start=12, tiles='CartoDB positron')

# 3. Add Network Connectivity (Weighted Edges)
for u, v, data in G_weighted.edges(data=True):
    pos_u = [hubs[u][0], hubs[u][1]]
    pos_v = [hubs[v][0], hubs[v][1]]
    
    folium.PolyLine(
        locations=[pos_u, pos_v],
        color='#2c3e50',
        weight=2,
        opacity=0.4,
        tooltip=f"Distance: {data['weight']:.2f} km"
    ).add_to(mapa_caba)

# 4. Add current taxi stops (Blue Points - Demand)
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=2,
        color='#3498db',
        fill=True,
        fill_color='#3498db',
        fill_opacity=0.4,
        popup=f"District: {row['barrio']}"
    ).add_to(mapa_caba)

# 5. Add Expansion Hubs (Proportional Red Circles - Supply)
for i, hub in enumerate(hubs):
    importance = centrality[i]
    # Radius grows based on centrality
    dynamic_radius = 6 + (importance * 40)
    
    folium.CircleMarker(
        location=[hub[0], hub[1]],
        radius=dynamic_radius,
        color='#e74c3c',
        fill=True,
        fill_color='#e74c3c',
        fill_opacity=0.8,
        popup=f"Hub {i+1}<br>Importance: {importance:.4f}"
    ).add_to(mapa_caba)

# 6. Strategic Legend
legend_html = """
<div style="
    position: fixed; 
    bottom: 30px; right: 30px; width: 260px; height: auto; 
    background-color: white; border: 1px solid #d1d1d1; z-index: 9999; 
    font-family: 'Arial', sans-serif; font-size: 13px; color: #333;
    box-shadow: 2px 2px 10px rgba(0,0,0,0.1); border-radius: 8px; 
    padding: 15px; line-height: 1.6;
">
    <b style="font-size: 14px; color: #000;">Advanced Strategic Roadmap</b><br>
    <hr style="margin: 8px 0; border: 0; border-top: 1px solid #eee;">
    <div style="display: flex; align-items: center; margin-bottom: 5px;">
        <span style="background: #3498db; border-radius: 50%; width: 12px; height: 12px; display: inline-block; margin-right: 10px;"></span>
        Taxi Stops (Market Demand)
    </div>
    <div style="display: flex; align-items: center; margin-bottom: 5px;">
        <span style="background: #e74c3c; border-radius: 50%; width: 14px; height: 14px; display: inline-block; margin-right: 10px;"></span>
        Expansion Hubs (Importance-sized)
    </div>
    <div style="display: flex; align-items: center;">
        <span style="background: #2c3e50; width: 20px; height: 2px; display: inline-block; margin-right: 10px;"></span>
        Supply Network (Weighted)
    </div>
    <div style="margin-top: 10px; font-size: 11px; color: #777; font-style: italic;">
        Optimized via Haversine & Centrality Metrics
    </div>
</div>
"""

mapa_caba.get_root().html.add_child(Element(legend_html))

# Save and Show
mapa_caba.save('strategic_network_analysis.html')
mapa_caba
```

## 6. Minimum Spanning Tree (MST): Kruskal’s Algorithm
To optimize the expansion further, we apply **Kruskal’s Algorithm**. This allows us to identify the "Minimum Essential Backbone" of our network.

- **Objective:** Connect all strategic Hubs using the minimum total distance possible.
- **Operational Value:** This represents the most cost-effective way to link all expansion zones, eliminating redundant paths and focusing on the core connectivity skeleton of the city.
- **Logic:** The algorithm sorts all possible connections by distance and adds them one by one, ensuring no loops (cycles) are formed, until every hub is part of a single unified tree.

```python
# 1. Calculate the MST using Kruskal's logic via NetworkX
# We use 'weight' (Haversine distance in km) to minimize total length
mst_graph = nx.minimum_spanning_tree(G_weighted, algorithm='kruskal', weight='weight')

# 2. Calculate Total Network Length
total_length = sum(d['weight'] for u, v, d in mst_graph.edges(data=True))

# 3. Create the Map
mapa_mst = folium.Map(location=[-34.6037, -58.3816], zoom_start=12, tiles='CartoDB positron')

# 4. Add MST Edges (The "Essential" Connections)
for u, v, data in mst_graph.edges(data=True):
    folium.PolyLine(
        locations=[[hubs[u][0], hubs[u][1]], [hubs[v][0], hubs[v][1]]],
        color='#27ae60', # Green to represent optimization/efficiency
        weight=3,
        opacity=0.8,
        tooltip=f"Essential Link: {data['weight']:.2f} km"
    ).add_to(mapa_mst)

# 5. Add Hubs (Using the Centrality size from before)
for i, hub in enumerate(hubs):
    importance = centrality[i]
    dynamic_radius = 6 + (importance * 40)
    
    folium.CircleMarker(
        location=[hub[0], hub[1]],
        radius=dynamic_radius,
        color='#e74c3c',
        fill=True,
        fill_color='#e74c3c',
        fill_opacity=0.8,
        popup=f"Hub {i+1}<br>Centrality: {importance:.4f}"
    ).add_to(mapa_mst)

# 6. Strategic Legend (MST focus)
legend_html = f"""
<div style="
    position: fixed; 
    bottom: 30px; right: 30px; width: 260px; height: auto; 
    background-color: white; border: 1px solid #d1d1d1; z-index: 9999; 
    font-family: 'Arial', sans-serif; font-size: 13px; color: #333;
    box-shadow: 2px 2px 10px rgba(0,0,0,0.1); border-radius: 8px; 
    padding: 15px; line-height: 1.6;
">
    <b style="font-size: 14px; color: #000;">Minimum Spanning Tree</b><br>
    <hr style="margin: 8px 0; border: 0; border-top: 1px solid #eee;">
    <div style="display: flex; align-items: center; margin-bottom: 5px;">
        <span style="background: #27ae60; width: 20px; height: 3px; display: inline-block; margin-right: 10px;"></span>
        Essential Infrastructure (MST)
    </div>
    <div style="display: flex; align-items: center; margin-bottom: 5px;">
        <span style="background: #e74c3c; border-radius: 50%; width: 12px; height: 12px; display: inline-block; margin-right: 10px;"></span>
        Network Hubs
    </div>
    <div style="margin-top: 10px; font-size: 11px; color: #777; font-style: italic;">
        Total Network Length: {total_length:.2f} km<br>
        Algorithm: Kruskal's MST
    </div>
</div>
"""

mapa_mst.get_root().html.add_child(Element(legend_html))

# Save and show
mapa_mst.save('strategic_mst_backbone.html')
mapa_mst
```

## 7. Strategic Conclusion: Optimizing Infrastructure Costs with MST

The implementation of **Kruskal’s Minimum Spanning Tree (MST)** provides the final layer of our urban mobility strategy. While Clustering identified demand centers and Centrality identified traffic pivots, the MST defines the **minimum essential architecture** for the network.

### Key Operational Takeaways:

1. **Cost-Effective Connectivity:** The MST identifies the set of paths that connect all strategic hubs with the **lowest total distance**. This represents the most efficient way to establish dedicated communication lines, charging infrastructure, or shuttle routes between hubs, avoiding redundant "cycles" or unnecessary investments.

2. **Differentiating "Essential" vs. "Optimal" Paths:**
   Unlike the full connectivity graph which shows all possible routes, the MST isolates the **critical skeleton** of the city. This allows the operations team to prioritize maintenance and driver incentives on the most vital "green" corridors shown in the map.

3. **Validation of Market Outliers:**
   The fact that certain hubs remain disconnected even in the MST confirms that these zones are **Satellite Markets**. This validates that our 5km distance threshold is correctly identifying urban areas that are geographically decoupled from the central CABA network.

### Final Summary of the Pipeline:
* **K-Means:** Solved the *Location* problem (Where should we be?).
* **Betweenness Centrality:** Solved the *Importance* problem (Which hubs are critical bridges?).
* **Kruskal’s MST:** Solved the *Cost* problem (How do we connect the city with minimum investment?).

This integrated approach combines **Unsupervised Learning** and **Operations Research** to transform raw taxi data into a high-level expansion roadmap.
